# 테스트 항목
## 제로샷(Zero-shot) 분류 성능
- 별도의 학습 없이 CLIP이 여러분의 상품 카테고리를 얼마나 잘 구별하는지 테스트합니다.

## 임베딩(Embedding) 추출 및 차원 분석
- 이미지와 텍스트를 벡터로 변환했을 때의 데이터 크기를 확인합니다.
- 추후 구축할 벡터 데이터베이스(Milvus, Pinecone 등)의 용량 산정과 직결

## 모델 사이즈별 성능/메모리 비교
- 모델을 각각 로드하여 VRAM 점유율과 정확도 사이의 트레이드오프를 측정.

## 추론 지연 시간(Latency)
- 단일 이미지 검색 시 소요되는 시간을 측정하여, 실제 서비스에서 사용자에게 제공할 수 있는 응답 속도의 기준점 수립.
